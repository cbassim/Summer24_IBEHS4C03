# Multiple Linear Regression with Ordinary Least Squares Methodologies with Python

```python
%reset
import pandas as pd
import numpy as np
```

```python
hypertension=pd.read_csv('Hypertension.csv')
hypertension
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig=plt.figure()
ax=fig.add_subplot(111)

sns.regplot(x='Age',y='BP',data=hypertension,ci=None)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)

sns.regplot(x='Weight',y='BP',data=hypertension,ci=None)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)

sns.regplot(x='Dur',y='BP',data=hypertension,ci=None)
```

```python
from statsmodels.formula.api import ols

model1=ols('BP~Age',data=hypertension)
results1=model1.fit()
print(results1.summary2())
```

```python
model2=ols('BP~Weight',data=hypertension)
results2=model2.fit()
print(results2.summary2())
```

```python
model3=ols('BP~Dur',data=hypertension)
results3=model3.fit()
print(results3.summary2())
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)

sns.residplot(x='Dur',y='BP',data=hypertension)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)

sns.scatterplot(x=hypertension["Weight"],y=results1.resid)
plt.hlines(0,86,102,color='red')
plt.ylabel("resdiuals from BP~Age")
```

```python
model4=ols('BP~Age+Weight',data=hypertension)
results4=model4.fit()
print(results4.summary2())
```

```python
sns.scatterplot(x=hypertension["Dur"],y=results4.resid)
plt.hlines(0,2,11,color='red')
plt.ylabel("resdiuals from BP~Age+Weight")
```

# Completion Time of a Maze

```python
time=pd.read_csv('CompletionTime.csv')
time
```

```python
sns.pairplot(time,kind='reg',diag_kind='hist')
```

```python
# invoke a corr() on the pandas DataFrame of time to determine the correaltions between the variables
correlations=time.corr()
correlations
```

```python
correlations=time.corr()
cmap=sns.color_palette("mako",as_cmap=True)

fig=plt.figure()
ax=fig.add_subplot(111)

sns.heatmap(correlations,annot=True,cmap=cmap)
```

```python
model5CT=ols('Completion~Brain',data=time)
results5CT=model5CT.fit()
print(results5CT.summary2())
```

```python
model5CTA=ols('Completion~Brain+Height',data=time)
results5CTA=model5CTA.fit()
print(results5CTA.summary2())
```

```python
%matplotlib widget
from mpl_toolkits.mplot3d import Axes3D

fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')

x=time['Brain']
y=time['Height']
z=time['Completion']
plt.xlabel("Brain")
plt.ylabel("Height")

ax.scatter(x,y,z)
ax.view_init(elev=20.,azim=20)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')

x=time['Brain']
y=time['Height']
z=time['Completion']
plt.xlabel("Brain")
plt.ylabel("Height")

ax.scatter(x,y,z)
ax.view_init(elev=20.,azim=20)

ax.scatter(x.values,y.values,results5CTA.fittedvalues.values,label="Fited Values")

x1,y1=model5CTA.exog_names[1:]

x_surf=np.arange(time['Brain'].min(), time['Brain'].max())
y_surf=np.arange(time['Height'].min(), time['Height'].max())
xS,yS=np.meshgrid(x_surf,y_surf)

exog=pd.DataFrame({x1: xS.ravel(), y1:yS.ravel()})
zS=results5CTA.predict(exog=exog).values.reshape(xS.shape)
ax.plot_surface(xS,yS,zS,alpha=0.4,color='orange')
```

```python
model5=ols('Completion~Brain+Height+Weight',data=time)
results5=model5.fit()
print(results5.summary2())
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)

sns.scatterplot(x=results5.fittedvalues.values,y=results5.resid)
plt.xlabel("fitted values")
plt.ylabel("residuals")
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm

fig=plt.figure(figsize=(12,8))
fig=sm.graphics.plot_regress_exog(results5,'Height',fig=fig)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)

sm.qqplot(results5.resid,line='s')
plt.show()
```

```python
import scipy
from scipy import stats

stats.shapiro(results5.resid)
```

```python
#Model building
#model5A=ols('Completion~Weight',data=time)
model5A=ols('Completion~Brain*Height',data=time)
results5A=model5A.fit()
print(results5A.summary2())
```

# MLR and Babies1

```python
babies=pd.read_csv('Babies1.csv')
babies
```

```python
sns.regplot(x="gestage",y="headcirc",data=babies)
```

```python
model6=ols('headcirc~gestage',data=babies)
results6=model6.fit()
print(results6.summary2())
```

```python
fig=plt.figure(figsize=(12,8))
fig=sm.graphics.plot_regress_exog(results6,'gestage',fig=fig)
```

```python
sm.qqplot(results6.resid,line='s')
plt.show()
```

```python
from statsmodels.stats.anova import anova_lm
anova_results6 = anova_lm(results6,type=3)
print(anova_results6)
```
