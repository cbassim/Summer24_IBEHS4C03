# Correlation and heatmaps with Python

In this notebook we will go over an example of how to perform a correlation analysis and then visualize a heatmap showing the linear relationship between many variables.

As usual we start by importing the data.

```python
%reset
import pandas as pd
import numpy as np
```

```python
Steel=pd.read_csv('Dilute_solyte_Diffusion.csv')
Steel
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.pairplot(Steel.iloc[:,2:10],diag_kind='kde')
```

```python
correlation_matrix=Steel.corr()
correlation_matrix
```

```python
sns.set(rc={"figure.figsize":(20, 20)})
sns.heatmap(correlation_matrix,annot=True)
```
